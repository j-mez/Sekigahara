```python
#Block stuff
class Block(object):
    color = ""
    clan = ""
    impact = 0
    b_type = ""

    # The class "constructor" - It's actually an initializer 
    def __init__(self, color, clan, impact, b_type):
        self.color = color
        self.clan = clan
        self.impact = impact
        self.b_type = b_type
    
    def info_self(self):
        if self.b_type=="Leader":
            symbol_rep = "üö©"
        elif self.b_type=="Army":
            symbol_rep = "‚öîÔ∏è"*self.impact #‚òÖ
        elif self.b_type=="Cavalry":
            symbol_rep = "üêé"
        elif self.b_type=="Arquebus":
            symbol_rep = "üî´"
        else:
            symbol_rep = "‚òÖ"
        return "{} {}({})".format(self.clan.upper(),symbol_rep,self.impact)
        
class BlockStack(object):
    block_array = []
    color = ""
    def __init__(self,block_array,color):
        self.block_array = block_array
        self.color = color
        
    def remove_block(self,block_index):
        '''
        Returns removed block
        '''
        return self.block_array.pop(block_index)
    
#     def combine_stacks(other_BlockStack):
#         '''
#         Returns new BlockStack with combined units
#         WILL NEED TO REMOVE OTHER BLOCKSTACK WHEN THEY COMBINE
        
#         '''
    def add_block(self,new_block):
        self.block_array.append(new_block)
        return True
    
    def split_stack(self,block_inds):
        '''
        Splits BlockStack into 2 stacks
        Returns two BlockStacks
        '''
        
    def sort_by_clan(self):
        '''
        Sorts blocks by clan
        '''
        self.block_array.sort(key=lambda block: block.clan, reverse=True)
        
    def display_stack_self(self):
        self.sort_by_clan()
        print("{} Stack, Size: {}".format(self.color,len(self.block_array)))
        index=0
        for block in self.block_array:
            print("{}. {}".format(index, block.info_self()))
            index+=1
        return True
    
    def display_block_other(self):
        print("{} {} Blocks".format(len(self.block_array),color))
    
```
